## Import necessary modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import load_breast_cancer

sns.set_style('whitegrid')
%matplotlib inline
```

### Load the dataset

```python
data_bc = load_breast_cancer()
```

```python
data_bc.keys()
```

```python
df = pd.DataFrame(data_bc['data'], columns=data_bc['feature_names'])
```

```python
df.head()
```

### Visualize as heat map

```python
plt.figure(figsize=(12,6))
sns.heatmap(df, cmap='viridis')
```

### Standardize the Scale

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

### Converting the scaled data to Dataframe

```python
scaled_df = pd.DataFrame(scaled_data, columns=data_bc['feature_names'])
```

```python
scaled_df.head()
```

### Visualize as heat map

```python
plt.figure(figsize=(12, 6))
sns.heatmap(scaled_df, cmap='viridis')
```

### Perform Principal Component Analysis to find the most useful feature

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_df)
```

```python
pca_arr = pca.transform(scaled_df)
```

### Visualize the different components

```python
sns.scatterplot(x=pca_arr[:, 0], y=pca_arr[:, 1], hue=data_bc['target'], palette='plasma')
```

```python
pca.components_
```

### Coverting the PCA components data to Dataframe

```python
pca_df = pd.DataFrame(pca.components_, columns=data_bc['feature_names'])
```

```python
pca_df.head()
```

```python
### Visualize as heat map
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(pca_df, cmap='viridis')
```

On seeing the heat map, mean fractal dimension, fractal dimension error and worst fractal dimension seems to be most useful feature
